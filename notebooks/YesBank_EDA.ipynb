```python
### Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Load Dataset
df = pd.read_csv('../data/data_YesBank_StockPrices.csv')
df.head()
```

```python
### Dataset Information
df.info()
```

```python
### Statistical Summary
df.describe()
```

```python
### Missing Values
df.isnull().sum()
```

```python
### Convert Date Column
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')
```

```python
### Correlation Matrix
corr = df.corr(numeric_only=True)
print(corr)
```

```python
### 1. Closing Price Trend
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'])
plt.title('Closing Price Trend')
plt.show()
```

```python
### 2. Opening Price Trend
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Open'])
plt.title('Opening Price Trend')
plt.show()
```

```python
### 3. High Price Trend
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['High'])
plt.title('High Price Trend')
plt.show()
```

```python
### 4. Low Price Trend
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Low'])
plt.title('Low Price Trend')
plt.show()
```

```python
### 5. Histogram of Close Price
plt.hist(df['Close'], bins=20)
plt.title('Histogram of Close Price')
plt.show()
```

```python
### 6. Histogram of Open Price
plt.hist(df['Open'], bins=20)
plt.title('Histogram of Open Price')
plt.show()
```

```python
### 7. Boxplot of Close Price
plt.boxplot(df['Close'])
plt.title('Boxplot of Close Price')
plt.show()
```

```python
### 8. Boxplot of Open Price
plt.boxplot(df['Open'])
plt.title('Boxplot of Open Price')
plt.show()
```

```python
### 9. Scatter Plot Open vs Close
plt.scatter(df['Open'], df['Close'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open vs Close')
plt.show()
```

```python
### 10. Scatter Plot High vs Close
plt.scatter(df['High'], df['Close'])
plt.xlabel('High')
plt.ylabel('Close')
plt.title('High vs Close')
plt.show()
```

```python
### 11. Scatter Plot Low vs Close
plt.scatter(df['Low'], df['Close'])
plt.xlabel('Low')
plt.ylabel('Close')
plt.title('Low vs Close')
plt.show()
```

```python
### 12. Correlation Heatmap
import seaborn as sns

plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
### 13. Moving Average Graph
df['MA_10'] = df['Close'].rolling(window=10).mean()

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], df['MA_10'], label='Moving Average')
plt.legend()
plt.show()
```

```python
### 14. Area Plot
df[['Open','Close']].plot.area(figsize=(10,5))
plt.show()
```

```python
### 15. Density Plot
df['Close'].plot(kind='density')
plt.title('Density Plot of Close Price')
plt.show()
```
